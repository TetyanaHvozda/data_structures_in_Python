2.2  Revise the charge and make_payment
methods of the CreditCard class to ensure that the caller sends a number
as a parameter.

```python
class CreditCard:
    def __init__(self, customer, bank, acnt, limit):
        self._customer = customer
        self._bank = bank
        self._account = acnt
        self._limit = limit
        self._balance = 0

    def get_customer(self):
        return self._customer

    def get_bank(self):
        return self._bank

    def get_account(self):
        return self._account

    def get_limit(self):
        return self._limit

    def get_balance(self):
        return self._balance

    def charge(self, price):
        if not isinstance(price, (int, float)):
            raise TypeError("Price must be a number")
        if price + self._balance > self._limit:
            return False
        else:
            self._balance += price
            return True

    def make_payment(self, amount):
        if not isinstance(amount, (int, float)):
            raise TypeError("Amount must be a number")
        self._balance -= amount
```

2.5 Implement the `__sub__` method for the Vector class, so
that the expression u−v returns a new vector instance representing the
difference between two vectors.

```python
class Vector:
    def __init__(self, d):
        if isinstance(d, (list, tuple)):
            self._coords = list(d)
        else:
            self._coords = [0] * d #d-dimensional vector of zeros

    def __len__(self):
        return len(self._coords)

    def __sub__(self, other):
        """Return the difference of two vectors."""
        if len(self) != len(other):
            raise ValueError('dimensions must agree')

        result = Vector(len(self)) #start with vector of zeros
        for j in range(len(self)):
            result._coords[j] = self._coords[j] - other._coords[j]
        return result
    
    # get clear, readable feedback
    def __repr__(self):
        return f"Vector({self._coords})"

# Create two 3D vectors
u = Vector([10, 20, 30])
v = Vector([1, 2, 3])

# Subtract v from u
w = u - v

print(u)
print(v)
print(w) # Output: Vector([9, 18, 27])
```

```python
class Vector:
    def __init__(self, d):
        if isinstance(d, (list, tuple)):
            self._coords = list(d)
        else:
            self._coords = [0] * d #d-dimensional vector of zeros

    def __len__(self):
        return len(self._coords)

    def __getitem__(self, j):
        return self._coords[j]

    def __setitem__(self, j, val):
        self._coords[j] = val

    def __add__(self, other):
        """Standard addition: Vector + Other"""
        if len(self) != len(other):
            raise ValueError('dimensions must agree')

        result = Vector(len(self)) #start with vector of zeros
        for j in range(len(self)):
            result[j] = self[j] + other[j]
        return result

    def __radd__(self, other):
        """Reflected addition: Other + Vector"""
        return self.__add__(other)
    
    # get clear, readable feedback
    def __repr__(self):
        return f"Vector({self._coords})"

# Create two 3D vectors
u = Vector([1, 2, 3, 4, 5])
v = u + [5, 3, 10,-2, 1]
w = [5, 3, 10,-2, 1] + u

print(u)
print(v)
print(w)
```

2.10 Modify the Sequence class below to include a definition for the `__eq__` method, so
that expression `seq1 == seq2` will `return True` precisely when the two
sequences are element by element equivalent.

```python
from abc import ABCMeta, abstractmethod

class Sequence(metaclass=ABCMeta):
    @abstractmethod
    def __len__(self):
        pass

    @abstractmethod
    def __getitem__(self, j):
        pass

    def __eq__(self, other):
        """Return True if two sequences are element-wise equivalent."""
        if len(self) != len(other):
            return False

        for j in range(len(self)):
            if self[j] != other[j]:
                return False

        return True

class MyList(Sequence):
    def __init__(self, data):
        self._data = data

    # If we don't implement __len__, Python throws an error
    def __len__(self):
        return len(self._data)

    def __getitem__(self, j):
        return self._data[j]

# This would FAIL because Sequence is abstract
# s = Sequence()

# This works!
m1 = MyList([1, 2, 3]) 
m2 = MyList([1, 2, 3]) 

print(m1 == m2)
```

2.11 Augment the Sequence class with
method `__lt__`, to support lexicographic comparison seq1 < seq2.

```python
from abc import ABCMeta, abstractmethod

class Sequence(metaclass=ABCMeta):
    @abstractmethod
    def __len__(self):
        pass

    @abstractmethod
    def __getitem__(self, j):
        pass

    def __lt__(self, other):
        """Return True if this sequence is lexicographically less than the other."""
        min_len = min(len(self), len(other))

        for j in range(min_len):
            if self[j] < other[j]:
                return True
            elif self[j] > other[j]:
                return False
                
        # If all compared elements are equal, the shorter sequence is 'less'
        return len(self) < len(other)

class MyList(Sequence):
    def __init__(self, data):
        self._data = data

    # If we don't implement __len__, Python throws an error
    def __len__(self):
        return len(self._data)

    def __getitem__(self, j):
        return self._data[j]
        
m1 = MyList([1, 2, 3]) 
m2 = MyList([1, 2, 3]) 
m3 = MyList([1, 2])
m4 = MyList([1, 3, 3])

print(f"m3 < m1:  {m3 < m1}")   # True (m3 is shorter but identical prefix)
print(f"m1 < m4:  {m1 < m4}")   # True (2 is less than 3 at index 1)
print(f"m4 < m1:  {m4 < m1}")   # False
print(f"m1 < m2:  {m1 < m2}")   # False (They are equal)
```

```python
import sympy as sp
from sympy.parsing.sympy_parser import parse_expr, standard_transformations, implicit_multiplication_application

def calculate_derivative():
    # Define 'x' as our mathematical symbol
    x = sp.symbols('x')

    print("Enter a polynomial in terms of 'x' (e.g., 3x^2 + 2x - 5):")
    poly_input = input("> ")

    try:
        # Pre-process: convert standard algebraic caret (^) to Python's power operator (**)
        poly_input_clean = poly_input.replace('^', '**')
        
        # Set up transformations to understand implicit multiplication (so '3x' is read as '3*x'), adding 2 tuples
        transformations = standard_transformations + (implicit_multiplication_application,)
        expr = parse_expr(poly_input_clean, transformations=transformations)
        derivative = sp.diff(expr, x)

        print(f"\nOriginal Polynomial: {expr}")
        print(f"First Derivative: {derivative}")
    except Exception as e:
        print(f"I couldn't parse that. Ensure it is a valid polynomial. (Error: {e})")

if __name__ == "__main__":
    calculate_derivative()
```

2.14 Write a set of Python classes that can simulate an Internet application in
which one party, Alice, is periodically creating a set of packets that she
wants to send to Bob. An Internet process is continually checking if Alice
has any packets to send, and if so, it delivers them to Bob’s computer, and
Bob is periodically checking if his computer has a packet from Alice, and,
if so, he reads and deletes it.

```python
# Data Objects (which hold information) and Behavioral Objects (which do the work)
import time
import random
from dataclasses import dataclass
from collections import deque

@dataclass
class Packet:
    packet_id: int
    sender: str
    receiver: str
    payload: str

class Host:
    """Represents a computer on the network with an inbox and outbox."""
    def __init__(self, name: str):
        self.name = name
        self.outbox = deque()
        self.inbox = deque()
        self._packet_counter = 0

    def create_packets(self, receiver: 'Host', num_packets: int):
        """Alice's action: Periodically creates packets to send."""
        for _ in range(num_packets):
            self._packet_counter += 1
            packet = Packet(
                packet_id=self._packet_counter,
                sender=self.name,
                receiver=receiver.name,
                payload=f"Message chunk #{self._packet_counter}"
            )
            self.outbox.append(packet)
            print(f"[{self.name}] created packet {packet.packet_id} for {receiver.name}.")

    def process_inbox(self):
        """Bob's action: Periodically checks, reads, and deletes packets."""
        if not self.inbox:
            print(f"[{self.name}] checked inbox, empty.")
            return
        while self.inbox:
            # popleft() simulates reading and deleting the oldest packet (FIFO)
            packet = self.inbox.popleft()
            print(f"[{self.name}] read packet {packet.packet_id} from {packet.sender}: '{packet.payload}'.")

# @staticmethod decorator, no need to pass self to this method. It doesn't need to know about the object. Acts like a standalone function
class Internet:
    """Represents the routing process checking outboxes and delivering to inboxes."""
    @staticmethod
    def route_packets(sender: Host, receiver: Host):
        if not sender.outbox:
            print("[Internet] Routing check: No packets waiting.")
            return

        while sender.outbox:
            packet = sender.outbox.popleft()
            receiver.inbox.append(packet)
            print(f"[Internet] delivered packet {packet.packet_id} from {sender.name} to {receiver.name}.")

def run_simulation(steps: int=5):
    alice = Host("Alice")
    bob = Host("Bob")
    network = Internet()

    for step in range(1, steps+1):
        print(f"Step {step}")
        
        #Alice periodically creates packets (randomly creates 0 to 3 packets)
        packets_to_create = random.randint(0, 3)
        if packets_to_create > 0:
            alice.create_packets(bob, packets_to_create)
        else:
            print(f"{alice.name} does not send anything.")

        # The Internet continually checks and routes
        network.route_packets(alice, bob)

        # Bob periodically checks his inbox
        bob.process_inbox()
        time.sleep(1) # Pause for 1 second to simulate time passing

if __name__ == "__main__":
    run_simulation()
```

2.15 Write a Python program to simulate an ecosystem containing two types
of creatures, bears and fish. The ecosystem consists of a river, which is
modeled as a relatively large list. Each element of the list should be a
Bear object, a Fish object, or None. In each time step, based on a random
process, each animal either attempts to move into an adjacent list location
or stay where it is. If two animals of the same type are about to collide in
the same cell, then they stay where they are, but they create a new instance
of that type of animal, which is placed in a random empty (i.e., previously
None) location in the list. If a bear and a fish collide, however, then the
fish dies (i.e., it disappears).

```python
import random

class Animal:
    def __repr__(self):
        return self.__class__.__name__[0]

class Bear(Animal):
    pass

class Fish(Animal):
    pass

class River:
    def __init__(self, size, num_bears, num_fish):
        self.grid = [None] * size
        self.size = size
        self.populate(num_bears, num_fish)

    def populate(self, bears, fish):
        """Initializes the river with a specific count of animals."""
        slots = random.sample(range(self.size), bears + fish)
        for i, slot in enumerate(slots):
            if i < bears:
                self.grid[slot] = Bear()
            else:
                self.grid[slot] = Fish()

    def get_empty_indices(self):
        return [i for i, val in enumerate(self.grid) if val is None]

    def add_random(self, animal_type):
        """Adds a new animal to a random empty spot if available."""
        empty = self.get_empty_indices()
        if empty:
            self.grid[random.choice(empty)] = animal_type()

    def step(self):
        indices = list(range(self.size))
        random.shuffle(indices)

        for i in indices:
            animal = self.grid[i]
            if animal is None:
                continue

            # Decide movement: -1 (left), 0 (stay), 1 (right)
            move = random.choice([-1, 0, 1])
            new_i = i + move

            if new_i < 0 or new_i >= self.size or move == 0:
                continue

            other = self.grid[new_i]

            if other is None:
                self.grid[new_i], self.grid[i] = self.grid[i], None
            elif type(animal) == type(other):
                self.add_random(type(animal))
            elif isinstance(animal, Bear) and isinstance(other, Fish):
                self.grid[new_i] = animal
                self.grid[i] = None
            elif isinstance(animal, Fish) and isinstance(other, Bear):
                self.grid[i] = None

    def display(self):
        print("".join(['.' if x is None else str(x) for x in self.grid]))

# --- Execution ---
river_size = 40
my_river = River(river_size, num_bears=5, num_fish=10)

print("Initial River (B=Bear, F=Fish, .=Empty):")
my_river.display()

for i in range(10):
    my_river.step()
    print(f"Step {i+1}:")
    my_river.display()
```

2.16 Write a Python program that simulates a system that supports the functions of an e-book reader. You should include methods for users of your
system to “buy” new books, view their list of purchased books, and read
their purchased books. Your system should use actual books, which have
expired copyrights and are available on the Internet, to populate your set
of available books for users of your system to “purchase” and read.

```python
import requests
import textwrap

class Book:
    def __init__(self, title, gutenberg_id):
        self.title = title
        self.id = gutenberg_id
        self.content = None

    def fetch_content(self):
        """Downloads the book text from Project Gutenberg."""
        if not self.content:
            url = f"https://www.gutenberg.org/cache/epub/{self.id}/pg{self.id}.txt"
            response = requests.get(url)
            if response.status_code == 200:
                self.content = response.text
            else:
                self.content = "Error: could not download book content."
        return self.content

class EReader:
    def __init__(self, username):
        self.username = username
        self.library = [] # purchased books
        self.store = {
            1: Book("Frankenstein", 84),
            2: Book("Pride and Prejudice", 1342),
            3: Book("Alice's Adventures in Wonderland", 11),
            4: Book("Metamorphosis", 5200)
        }
    def show_store(self):
        print(f"\n--- Welcome to the E-Book Store, {self.username}! ---")
        for key, book in self.store.items():
            status = "[Purchased]" if book in self.library else ""
            print(f"{key}. {book.title} {status}")

    def buy_book(self, book_id):
        book = self.store.get(book_id)
        if not book:
            print("Invalid Book ID.")
        elif book in self.library:
            print(f"You already own '{book.title}'.")
        else:
            self.library.append(book)
            print(f"Successfully purchased '{book.title}'!")

    def view_library(self):
        print(f"\n {self.username}'s Library")
        if not self.library:
            print("Your library is empty. Go to the store")
        else:
            for i, book in enumerate(self.library, 1):
                print(f"{i}. {book.title}")

    def read_book(self, library_index):
        if 0 <= library_index < len(self.library):
            book = self.library[library_index]
            print(f"\nOpening '{book.title}'...")
            text = book.fetch_content()

            preview = textwrap.fill(text[1000:2500], width=70)
            print("-" * 70)
            print(preview)
            print("-" * 70)
            print("... (End of Preview) ...")
        else:
            print("Invalid selection.")

reader = EReader("Alex")

reader.show_store()
reader.buy_book(1)
reader.buy_book(3)

reader.view_library()

reader.read_book(0)
```

```python
import math

class Polygon:
    def area(self):
        pass

    def perimeter(self):
        pass

class Triangle(Polygon):
    def __init__(self, a, b, c):
        self.sides = [a, b, c]

    def perimeter(self):
        return sum(self.sides)

    def area(self):
        # semi-perimeter.
        s = self.perimeter() / 2
        a, b, c = self.sides
        return math.sqrt(s * (s - a) * (s - b) * (s - c))

class Quadrilateral(Polygon):
    def __init__(self, sides):
        self.sides = sides #expects list of 4 sides
    def parameter(self):
        return sum(self.sides)

class Pentagon(Polygon):
    pass

class Hexagon(Polygon):
    pass

class Octagon(Polygon):
    pass

class IsoscelesTriangle(Triangle):
    pass

class EquilateralTriangle(Triangle):
    def __init__(self, side):
        super().__init__(side, side, side)

class Rectangle(Quadrilateral):
    def __init__(self, width, height):
        super().__init__([width, height, width, height])
        self.w = width
        self.h = height
    def area(self):
        return self.w * self.h

class Square(Rectangle):
    def __init__(self, side):
        super().__init__(side, side)

def main():
    print("1. Rectangle\n2. Equilateral Triangle\n3. Square")
    choice = input("Select a polygon: ")

    if choice == "1":
        w = float(input("Enter width: "))
        h = float(input("Enter height: "))
        poly = Rectangle(w, h)
    elif choice == "3":
        s = float(input("Enter side: "))
        poly = Square(s)

    print(f"Perimeter: {poly.perimeter()}")
    print(f"Area: {poly.area()}")

if __name__ == "__main__":
    main()
```
